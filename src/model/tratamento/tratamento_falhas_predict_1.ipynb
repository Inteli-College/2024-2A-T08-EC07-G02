## Base de dados "FALHAS_PREDICT_1"


```python
# Leitura da base de dados falhas
df_falhas = pd.read_csv('FALHAS_PREDICT_1.csv')
df_falhas
```

```python
# Dropando a coluna Unnamed
df_falhas.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Visualização da forma da base de dados para saber com que estamos lidando
df_falhas.shape
```

```python
# Para saber que os dados que estão dispostos para trabalharmos
df_falhas.columns
```

```python
# Saber os tipos de cada coluna para ter noção do que vamos lidar 
df_falhas.info()
```

```python
# Ter informações de quais dados são nulos e sua quantidade para saber quatos irão ser descartados
df_falhas.isna().sum()
```

```python
# Visualizar quais modelos estão dispostos na base de dados 
df_falhas['MODELO'].value_counts()
```

```python
# Limpar a base de dados para excluir todos os campos que estão vazios
df_falhas_limpa = df_falhas.dropna()
df_falhas_limpa.shape
```

Com a remoção dos dados vazios houve foi deletado 1862 dados da tabela

```python
# Ter mais informações sobre como os dados estao organizados em seus tipos
df_falhas_limpa.info()
```

```python
# Pegar somente os dados do tipo T-Cross, pois sera o foco da analise
df_falhas_limpa1 = df_falhas_limpa[df_falhas_limpa['MODELO'] == 'T-CROSS']
df_falhas_limpa1['MODELO'].unique()
```

```python
# Pegar somente os dados do tipo ROD, pois será o foco da análise
df_falhas_limpa = df_falhas_limpa1[df_falhas_limpa1['HALLE'] == 'ROD']

df_falhas_limpa
```

```python
# Visualizar quantos tipos de falhas diferentes há na base de dados
total_falhas = df_falhas_limpa['FALHA'].unique()
len(total_falhas)
```

```python
# Visualizar qual tipo de falha é a mais comum 
df_falhas_limpa['FALHA'].value_counts()
```

```python
# Dropando colunas que não serao utilizadas
df_falhas_limpa = df_falhas_limpa.drop(['MODELO','DATA'], axis='columns')
df_falhas_limpa
```

```python
# Coloca todas as letras em maiúscula para que possa pegar todos os dados repetidos
df_falhas_limpa['FALHA'] = df_falhas_limpa['FALHA'].str.upper()
df_falhas_limpa
```

```python
# Visualizando o total de falhas diferentes presentes na tabela

total_falhas = df_falhas_limpa['FALHA'].unique()
len(total_falhas)
```

```python
# Elimina os dados repetidos

df_falhas_limpa = df_falhas_limpa.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df_falhas_limpa = df_falhas_limpa.astype(str)
df_falhas_limpa = df_falhas_limpa.drop_duplicates()
```

```python
# Verificando o tamanho da base de dados
df_falhas_limpa.shape
```

```python
# Colunas que não serão utilizadas no modelo
lixo = {"COR", "MOTOR","ESTACAO", "USUARIO", "HALLE" }
df_falhas_limpa.drop(columns=lixo)

```

```python
# Criando uma nova tabela a partir da df_falhas_limpa
df_falhas_limpa_true = df_falhas_limpa
df_falhas_limpa_true
```

```python
# Preenchendo os resultados nulos por 0
df_falhas_limpa_true_filled = df_falhas_limpa_true.fillna(0)

# Transforma os valores não nulos para 1
df_falhas_limpa_true_filled[df_falhas_limpa_true_filled != 0] = 1
```

```python
# Apando colunas que não serão utilizada no modelo
apagar = {'COR', 'MOTOR', 'ESTACAO', 'USUARIO', 'HALLE'}
df_falhas_limpa_true = df_falhas_limpa_true.drop(columns=apagar)
```

```python
# Atribuindo para coluna FALHA todos os valores como 1 para indicar que há uma falha 
df_falhas_limpa_true['FALHA'] = 1
```

```python
# Eliminando KNRs repetidos
df_falhas_limpa_true_unicos = df_falhas_limpa_true.drop_duplicates(subset='KNR')
df_falhas_limpa_true_unicos2 = df_falhas_limpa_true_unicos.drop(columns=['S_GROUP_ID'])
df_falhas_limpa_true_unicos2
```

```python
# Salvando a tabela tratada
df_falhas_limpa_true_unicos2.to_csv('tabelas finais/FALHAS_PREDICT_1_TRUE.csv', index=False)
```
