```python
import pandas as pd
import numpy as np
import dask.dataframe as dd

BASE_PATH = 'temp/'
```

```python
FAILURES_PATH = f'{BASE_PATH}FAILURES.csv'
failures_df = pd.read_csv(FAILURES_PATH)

failures_df
```

```python
failures_df.to_parquet(f'{BASE_PATH}FAILURES.parquet', index=False)
failures_df = pd.read_parquet(f'{BASE_PATH}FAILURES.parquet')

failures_df
```

```python
RESULTS_PATH = f'{BASE_PATH}RESULTS.csv'

results_df = pd.DataFrame()

batch_size = 100000
current_chunk = 0
for chunk in pd.read_csv(RESULTS_PATH, chunksize=batch_size):
    current_chunk += 1
    print(f'Processing chunk {current_chunk}')
    chunk['ID'] = pd.to_numeric(chunk['ID'], downcast='integer')
    chunk['STATUS'] = pd.to_numeric(chunk['STATUS'], downcast='integer')
    chunk['VALUE_ID'] = pd.to_numeric(chunk['VALUE_ID'], downcast='integer', errors='coerce')
    chunk["VALUE"] = pd.to_numeric(chunk["VALUE"], errors='coerce')

    results_df = pd.concat([results_df, chunk], ignore_index=True)
    print(f'Chunk {current_chunk} processed, current size: {results_df.shape[0]}')
```

```python
RESULTS_PATH = f'{BASE_PATH}RESULTS.csv'

dtype = {
    'ID': 'int32',
    'STATUS': 'int8',
    'VALUE_ID': 'float32',
    'VALUE': 'float32',
    'NAME': 'object',
}

results_df = dd.read_csv(RESULTS_PATH, dtype=dtype, na_values=[' ', '', '#MULTIVALUE'],skipinitialspace=True)
results_df = results_df.compute()

results_df
```

```python
results_df.to_parquet(f'{BASE_PATH}RESULTS.parquet', index=False)
results_df = pd.read_parquet(f'{BASE_PATH}RESULTS.parquet')

results_df
```
