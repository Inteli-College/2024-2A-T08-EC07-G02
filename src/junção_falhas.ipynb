```python
import pandas as pd
import numpy as np
```

```python
# Definindo as tabelas
df_falhas_predict = pd.read_csv('tabelas finais/FALHAS_PREDICT_TRUE.csv')
df_falhas_predict_1 = pd.read_csv('tabelas finais/FALHAS_PREDICT_1_TRUE.csv')
df_falhas = pd.read_csv('tabelas finais/FALHAS_TRUE.csv')
```

```python
# Visualização do número de falhas
df_falhas['FALHA'].value_counts()
```

```python
# Juntando as bases de dados por KNR
quase_resultado = pd.concat([df_falhas_predict, df_falhas_predict_1, df_falhas])

quase_resultado
```

```python
# Retirando os valores duplicados
resultado = quase_resultado.drop_duplicates(subset='KNR')
resultado
```

```python
# Salvando a tabela final
resultado.to_csv('tabelas finais/PREDICTS_FINAIS.csv', index=False)
```
